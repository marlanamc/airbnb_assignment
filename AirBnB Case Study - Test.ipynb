```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df.info()
```

```python
# How many neighborhood groups are available and which shows up the most?
neighborhoods = df["neighbourhood_group"].value_counts()
neighborhoods
```

#### There are 5 neighborhood groups 

```python
# Are private rooms the most popular in manhattan?
manhattan = df[(df["neighbourhood_group"] == "Manhattan")]
manhattan
```

```python
manhattan_rooms = df["room_type"].value_counts(normalize=True)
manhattan_rooms
```

#### Entire homes/apartments are the most popular in Manhattan at 52% of listings

```python
# Which hosts are the busiest and based on their reviews?
review = df.sort_values('reviews_per_month',ascending=False)
review.head(5)
```

#### Row NYC, Louann, and Nalicia are very busy according to the amount of reviews per month

```python
#Which neighorhood group has the highest average price?
neigh = df.groupby("neighbourhood_group")["price"].mean().sort_values(ascending=False)
neigh
```

#### Manhattan has the highest average price at $196 per night 

```python
# Which neighbor hood group has the highest total price?
neigh_tot = df.groupby("neighbourhood_group")["price"].sum().sort_values(ascending=False)
#a = neigh_tot / neigh_tot
#a
neigh_tot
```

```python
#Which top 5 hosts have the highest total price?
top_hosts = df.groupby("host_name")["price"].sum().sort_values(ascending=False)
top_hosts
```

#### Hosts: Sonder (NYC), Blueground, Michael, David, Alex have the highest total prices total across all their listings 

```python
num_prop = df.groupby("host_name")["id"].nunique().sort_values(ascending=False)
num_prop
```

```python
b= top_hosts / num_prop
b.sort_values(ascending=False)
```

#### Olson lists the highest listing at $9999

```python
# Who currently has no (zero) availability with a review count of 100 or more?
review_100 = df[(df["number_of_reviews"] >= 100)]
review_100
```

```python
no_avail_review_100  = review_100[(review_100["availability_365"] == 0)]
no_avail_review_100
```

```python
no_avail_review_100_count  = review_100[(review_100["availability_365"] == 0)].nunique()
no_avail_review_100_count
```

#### 162 listings have 0 availability with a review count of over 100; 148 unique hosts
